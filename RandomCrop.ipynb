```python
import os
import pathlib
import tensorflow as tf
from os import listdir
from numpy import expand_dims
from os.path import join, isfile
from tensorflow.keras.preprocessing.image import load_img, ImageDataGenerator, img_to_array
```

```python
current_directory = os.getcwd()
print(current_directory)
```

```python


def random_crop(image):
    #print(image.size)
    height, width = image.size
    
    x = int(height/2)
    y = int(width/2)
    
    cropped_image = tf.image.random_crop(image, size=[224, 224, 3])
    #print(cropped_image.shape)
    return cropped_image

def generateRandomCropFromDirectory(dir, save_dir):

    files = [f for f in listdir(dir) if isfile(join(dir, f))]
    
    for file in files:
      if (file.lower().endswith(('.jpg'))):

        img_path = dir + file        
        img = load_img(img_path)        
        img = random_crop(img)        
        img_data = img_to_array(img)
        img_sample = expand_dims(img_data, 0)
        random_crop_in_datagen = ImageDataGenerator(zoom_range=0)
        #zoo igual a zero é só uma gambiarra pra rodar o codigo        
        filename = file.replace(".jpg", "")
        print(save_dir)
        for batch in random_crop_in_datagen.flow(img_sample, 
            batch_size=1,
            save_to_dir=save_dir, 
            save_prefix=filename, 
            save_format='jpg'):
            break
                       
RandomUnderNormal_save_dir = current_directory + "\\InterestRegion\\Normal\\22Under\\RandomCrop/"
RandomOverNormal_save_dir =current_directory + "\\InterestRegion\\Normal\\22OrOver\\RandomCrop/"

RandomUnderLoss_save_dir = current_directory +"\\InterestRegion\\Resolution-Loss\\22Under\\RandomCrop/"
RandomOverLoss_save_dir = current_directory +"\\InterestRegion\\Resolution-Loss\\22OrOver\\RandomCrop/"



data22UnderRawNormal_dir = current_directory + "\\InterestRegion\\Normal\\22Under\\Raw/"
data22UnderZoomNormal_dir = current_directory + "\\InterestRegion\\Normal\\22Under\\Zoom/"
data22UnderRotationNormal_dir = current_directory + "\\InterestRegion\\Normal\\22Under\\Rotation/"


data22OverRawNormal_dir = current_directory + "\\InterestRegion\\Normal\\22OrOver\\Raw/"
data22OverZoomNormal_dir = current_directory + "\\InterestRegion\\Normal\\22OrOver\\Zoom/"
data22OverRotationNormal_dir = current_directory + "\\InterestRegion\\Normal\\22OrOver\\Rotation/"


# ##############################################################################
data22UnderRawLoss_dir = current_directory + "\\InterestRegion\\Resolution-Loss\\22Under\\Raw/"
data22UnderZoomLoss_dir = current_directory + "\\InterestRegion\\Resolution-Loss\\22Under\\Zoom/"
data22UnderRotationLoss_dir = current_directory + "\\InterestRegion\\Resolution-Loss\\22Under\\Rotation/"


data22OverRawLoss_dir = current_directory + "\\InterestRegion\\Resolution-Loss\\22OrOver\\Raw/"
data22OverZoomLoss_dir = current_directory + "\\InterestRegion\\Resolution-Loss\\22OrOver\\Zoom/"
data22OverRotationLoss_dir = current_directory + "\\InterestRegion\\Resolution-Loss\\22OrOver\\Rotation/"
# ################################################################################
generateRandomCropFromDirectory(data22UnderRawNormal_dir, RandomUnderNormal_save_dir)
generateRandomCropFromDirectory(data22UnderZoomNormal_dir, RandomUnderNormal_save_dir)
generateRandomCropFromDirectory(data22UnderRotationNormal_dir, RandomUnderNormal_save_dir)

generateRandomCropFromDirectory(data22OverRawNormal_dir, RandomOverNormal_save_dir)
generateRandomCropFromDirectory(data22OverZoomNormal_dir, RandomOverNormal_save_dir)
generateRandomCropFromDirectory(data22OverRotationNormal_dir, RandomOverNormal_save_dir)

# ###################################################################################3

generateRandomCropFromDirectory(data22UnderRawLoss_dir, RandomUnderLoss_save_dir)
generateRandomCropFromDirectory(data22UnderZoomLoss_dir, RandomUnderLoss_save_dir)
generateRandomCropFromDirectory(data22UnderRotationLoss_dir, RandomUnderLoss_save_dir)

generateRandomCropFromDirectory(data22OverRawLoss_dir, RandomOverLoss_save_dir)
generateRandomCropFromDirectory(data22OverZoomLoss_dir, RandomOverLoss_save_dir)
generateRandomCropFromDirectory(data22OverRotationLoss_dir, RandomOverLoss_save_dir)


```
