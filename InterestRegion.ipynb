```python
import os
import cv2
from numpy import expand_dims
from matplotlib import pyplot
from tensorflow.keras.preprocessing.image import load_img, ImageDataGenerator, img_to_array
from tensorflow.data import Dataset
import pathlib
from os import listdir
from os.path import join, isfile
import tensorflow as tf

def generateInterestRegionAreaFromDirectory(dir, save_dir):

    files = [f for f in listdir(dir) if isfile(join(dir, f))]
    
    for file in files:
      if (file.lower().endswith(('.jpg'))):

        img_path = str(dir.resolve()) + "/" +  file
        img = load_img(img_path)
        # offset_height = 600
        # offset_width = 650
        # target_height = 700
        # target_width = 1500

        # offset_height = 400
        # offset_width = 500
        # target_height = 600
        # target_width = 1000

        offset_height = 400
        offset_width = 400
        target_height = 500
        target_width = 1000
        cropped_img = tf.image.crop_to_bounding_box(img, offset_height, offset_width, target_height, target_width)

        filename = file.replace(".jpg", "")
        save_path = save_dir + filename + ".jpg"
        tf.keras.utils.save_img(save_path, cropped_img.numpy())


```
